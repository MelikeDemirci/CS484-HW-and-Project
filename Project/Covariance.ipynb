```python
import numpy as np
from skimage.io import imread, imshow
from skimage.transform import resize
import tqdm
import matplotlib.pyplot as plt
```

```python
def Covariance(x, y):
    xbar, ybar = x.mean(), y.mean()
    return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)
```

```python
non1 = resize(imread('non1.jpg'), (128, 128, 3))
non2 = resize(imread('non2.jpg'), (128, 128, 3))
smoke1 = resize(imread('smoke1.jpg'), (128, 128, 3))
smoke2 = resize(imread('smoke2.jpg'), (128, 128, 3))

c1 = Covariance(non1,non1)
c2 = Covariance(smoke1,smoke2)
c3 = Covariance(non1,smoke1)
print( "Covariance of two fire images: %.4f" % c1)
print( "Covariance of two smoke images: %.4f" % c2)
print( "Covariance of fire and smoke images: %.4f" % c3)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].imshow(non1)
ax[1].imshow(non2)
plt.title("Covariance : " + str(c1))
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].imshow(smoke1)
ax[1].imshow(smoke2)
plt.title("Covariance : " + str(c2))
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].imshow(smoke1)
ax[1].imshow(non1)
plt.title("Covariance : " + str(c3))
```
